# 4.Добавление новых столбцов и строк
## Описание задания:
Используя Таблицу 1:
Добавьте новый столбец "Категория зарплаты" : "Низкая" – если Зарплата < 60000 ."Средняя" – если 60000 ≤ Зарплата < 100000 .

"Высокая" – если Зарплата ≥ 100000

```python
conditions = [
    (df_1['Зарплата'] < 60000),
    (df_1['Зарплата'] >= 60000) & (df_1['Зарплата'] < 100000),
    (df_1['Зарплата'] >= 100000)
]
categories = ["Низкая", "Средняя", "Высокая"]
df_1["Категория зарплаты"] = np.select(conditions, categories)
display(df_1)
```
